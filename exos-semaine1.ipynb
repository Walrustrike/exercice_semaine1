```python
from random import randrange


def cadre(mat):
    l,c = mat.shape
    for i in range(l):
        mat[i][1] = 0
        mat[i][c-2] = 0
    for j in range(c):
        mat[1][j] = 0
        mat[l-2][j] = 0
```

```python
c = np.ones((10,20))
cadre(c)
plt.imshow(c)
;
```

```python
tabrandom = np.random.randint(0, 21, (6,8), dtype = np.uint8)
plt.imshow(tabrandom)
;
```

```python
cadre(tabrandom)
plt.imshow(tabrandom)
;
```

Cela ne va pas fonctionner pour un tableau de trop petite taille

```python
taberreur = np.ones((2,1))
cadre(taberreur)
```

Cela ne va pas non plus fonctionner pour les matrices où le type des éléments n'est pas entier ou flottant

On ne peut pas faire de copie car on va modifier les dimensions du ndarray.

```python
def cadre2(mat):
    l,c = mat.shape
    mat2 = np.zeros((l+2,c+2))
    for i in range(l):
        for j in range(c):
            mat2[i+1][j+1] = mat[i][j]
    plt.imshow(mat2)

cadre2(tabrandom)
```

Cela assure la correction de l'algorithme, on ne rencontrera pas de problème de typage ou de taille

exo anti_diagonal

```python
def anti_diagonal(listenb):
    l = len(listenb)
    tab = np.zeros((l,l))
    for i in range(l):
        tab[i][l-i-1] = listenb[i]
    return tab

print(anti_diagonal([1,2,3,4]))
```
